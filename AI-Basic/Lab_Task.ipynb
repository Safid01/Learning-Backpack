```python
# Local Beam Search — Maximization

tree = {
    'S': (10,  ['A', 'B', 'C', 'D']),
    'A': (40,  []),
    'B': (70,  ['E', 'F']),
    'C': (55,  ['G']),
    'D': (30,  []),
    'E': (90,  []),
    'F': (65,  []),
    'G': (80,  []),
}

def local_beam_search(start, k):
    beam = [start]
    print(f"Local Beam Search | start={start}, Beam Width(k)={k}\n")

    step = 1
    while beam:
        print(f"Step {step} | Beam: {beam}")

        neighbors = [child for node in beam for child in tree[node][1]]

        if not neighbors:
            print("No neighbors found. Stopping.")
            break

        beam = sorted(neighbors, key=lambda n: tree[n][0], reverse=True)[:k]
        print(f"  All neighbors: {neighbors}")
        print(f"  Top {k} selected (Best First): {beam} (values: {[tree[n][0] for n in beam]})\n")
        step += 1

    best = max(beam, key=lambda n: tree[n][0])
    print(f"Best node: {best} | Value: {tree[best][0]}")

local_beam_search('S', k=2)
```

```python
# Simulated Annealing — Maximization
import math, random

tree = {
    'S': (100, ['A', 'B', 'C', 'D']),
    'A': (95,  []),
    'B': (105, []),
    'C': (90,  []),
    'D': (85,  []),
}

def simulated_annealing(start, T=10, cooling=1, max_iter=10):
    random.seed(42)
    current, value = start, tree[start][0]
    print(f"Simulated Annealing | start={current}, T={T}, cooling={cooling}\n")

    for i in range(max_iter):
        print(f"Iter {i+1} | T={T} | current={current} (value={value})")
        children = tree[current][1]

        if not children:
            print("  No children. Stopping.")
            break

        for child in children:
            delta = tree[child][0] - value
            prob = 1.0 if delta > 0 else math.exp(delta / T)
            r = random.uniform(0, 1)
            print(f"  Child {child}: delta={delta}, P={prob:.4f}, r={r:.4f} → {'Accept' if r < prob else 'Reject'}")

            if r < prob:
                current, value = child, tree[child][0]
                break

        T -= cooling
        if T <= 0:
            print("\nTemperature reached 0. Stopping.")
            break

    print(f"\nBest node: {current} | Heuristic Value: {value}")

simulated_annealing('S')
```

```python
# Hill Climbing — Minimization

tree = {
    'S': (200, ['A', 'B', 'C']),
    'A': (295, ['F', 'G']),
    'B': (290, []),
    'C': (195, ['D', 'E']),
    'F': (96,  ['H']),
    'G': (84,  ['I']),
    'D': (110, ['J']),
    'E': (85,  []),
    'H': (80,  []),
    'I': (109, []),
    'J': (128, []),
}

def hill_climbing(start):
    current = start
    print(f"Hill Climbing | start={current}\n")

    while True:
        value = tree[current][0]
        children = sorted(tree[current][1])
        print(f"Current: {current} (value={value}) | Children: {[(c, tree[c][0]) for c in children]}")

        if not children:
            print(f"  No children. Stopping.")
            break

        best = min(children, key=lambda n: tree[n][0])
        best_val = tree[best][0]

        if best_val < value:
            print(f"  Move to {best} (value={best_val}) ✅\n")
            current = best
        else:
            print(f"  No improvement. Local minimum reached.")
            break

    print(f"\nFinal: {current} | Heuristic: {tree[current][0]}")

hill_climbing('S')
```
