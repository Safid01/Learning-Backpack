# Uniform Cost Search

```python
from queue import PriorityQueue


def ucs(snode, dnode, GRAPH):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()

    node = (0,snode,[snode]) #(g(n) = path cost, city name, path)
    frontier.put(node)
    i = 1
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1]
        if u in expanded:
          continue
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == dnode:# goal test
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i +1
        for v in GRAPH[u].keys():#expanding
            if v not in expanded:
                g = node[0] + GRAPH[u][v]
                path = node[2] + [v]
                frontier.put((g,v,path))
                print(u + "->"+v + ", Path: "+ str(path) + ", Cost:"+ str(g))
        #print(frontier.queue)
    print('Failed')
```

```python
ucs('S', 'G', EX1) # change source and destination
```

# Greedy Best First Search

```python
h1 = {'S':10,'A':8, 'B':5, 'C':5,
     'D':4,'E':2, 'G':0}
```

```python
from queue import PriorityQueue

def gbfs(snode, dnode, GRAPH, heur):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()

    snode = (heur[snode],snode,[snode],0) # (h, city, path, path cost)
    frontier.put(snode)
    i =1
    while not frontier.empty():
        node = frontier.get()
        u = node[1]
        if u in expanded:
          continue

        print(u+ " selected. Cost: "+ str(node[0]))
        if u== dnode:#Goal Test
            print(node)
            print(expanded)
            return node #path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i+1
        for v in GRAPH[u].keys():
            if v not in expanded:
                g = node[3] + GRAPH[u][v]
                path = node[2]+[v]
                frontier.put((heur[v],v,path, g))
                print(u + "->"+v + ", Path: "+ str(path) + ", h:"+ str(heur[v]))
    print('Failed')
```

```python
gbfs('S', 'G', EX1, h1) # change source only
```

# A*

```python
import heapq

def astar(graph, start, goal, heuristic):

    # Priority queue
    open_list = []
    heapq.heappush(open_list, (0, start))

    # Cost from start
    g_cost = {node: float('inf') for node in graph}
    g_cost[start] = 0

    # Parent tracking
    parent = {start: None}

    while open_list:

        current = heapq.heappop(open_list)[1]

        if current == goal:
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]

        for neighbor, cost in graph[current].items():

            new_cost = g_cost[current] + cost

            if new_cost < g_cost[neighbor]:
                g_cost[neighbor] = new_cost
                f_cost = new_cost + heuristic[neighbor]

                heapq.heappush(open_list, (f_cost, neighbor))
                parent[neighbor] = current

    return None
```

```python
start = 'S'
goal = 'G'

path = astar(EX1, start, goal, h1)

print("Shortest Path:", path)
```
