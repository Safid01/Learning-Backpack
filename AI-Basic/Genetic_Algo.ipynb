```python
import random
def initPopulation(b, n):
  p = [random.choices([0,1],k = b) for i in range(n)]
  return p
```

```python
#Population Initialization
initPopulation(8,12)
```

```python
# Parent selection
def binToDecimal(binList):
  decValue = 0
  power = len(binList) - 1
  for digit in binList:
    decValue += digit * 2 ** power
    power -= 1
  return decValue

def func(x):
  return x**2 - 11*x + 150

def getFitness(p):
  f = []
  for sol in p:
    d = binToDecimal(sol)
    v = func(d)
    if v < 0: f.append(0)
    else: f.append(v)
  return f
```

```python
# Calculate probability and select parents
import numpy as np
def select_parent(n, pop):
    fitness = getFitness(pop)
    total_fitness = sum(fitness)
    prob = [f/total_fitness for f in fitness]
    cumSumProb = np.cumsum(prob)
    parents = []
    for i in range(n):
        r = random.random()
        for j in range(len(cumSumProb)):
            if r <= cumSumProb[j]:
                parents.append(pop[j])
                break
    return parents
```

```python
population = initPopulation(8,10)
print(population)
parents = select_parent(4, population)
print(parents)
```

```python
import random
# Crossover
def cross(p):
  num_cross = len(p)//2
  offsprings = []
  for i in range(num_cross):
   cp = random.choice(range(3,6))
   x,y = p[0], p[1]
   off1 = x[:cp] + y[cp:]
   offsprings.append(off1)

   off2 = y[:cp] + x[cp:] # Code for off2
   offsprings.append(off2)

  return offsprings
```

```python
#Select survivors
def select_survivor(offsp):
  topOffsp = []
  f = getFitness(offsp)
  mInd = f.index(max(f))
  topOffsp.append(offsp[mInd])
  f.pop(mInd)
  #print(f)
  mInd = f.index(max(f))
  topOffsp.append(offsp[mInd])
  return topOffsp
```

```python
def best(p):
  f = getFitness(p)
  mInd = f.index(max(f))
  return p[mInd], max(f)
```

```python
# Iteration
population = initPopulation(8,10)
print(best(population))
parents = select_parent(6, population)
offsprings = cross(parents)
# Call mutation here
survivors = select_survivor(offsprings)
updatedPopulation = population + survivors
updatedPopulation
```
