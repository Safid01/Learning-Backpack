```python
import random
def initPopulation(b, n):
  p = [random.choices([0,1],k = b) for i in range(n)]
  return p
```

```python
#Intialization
initPopulation(7,5)
```

```python
def binToDecimal(binList):
  decValue = 0
  power = len(binList) - 1
  for digit in binList:
    decValue += digit * 2 ** power
    power -= 1
  return decValue

def func(x):
  return (2*x**3)/3 - 2*x + 12

def getFitness(p):
  f = []
  for sol in p:
    d = binToDecimal(sol)
    v = func(d)
    if v < 0: f.append(0)
    else: f.append(v)
  return f
```

```python
population = initPopulation(7,5)
getFitness(population)
```

```python
import numpy as np
def select_parent(pop):
    fitness = getFitness(pop)
    indexed = sorted(enumerate(fitness), key=lambda t: t[1], reverse=True)
    return pop[indexed[0][0]], pop[indexed[1][0]]
```

```python
population = initPopulation(7,5)
print(population)
parents = select_parent(population)
print(parents)
```

```python
import random
def cross(p):
  cp = 2
  x, y = p[0], p[1]
  off1 = x[:cp] + y[cp:]
  off2 = y[:cp] + x[cp:]
  return [off1, off2]
```

```python
def mutate(offsp):
  for child in offsp[:2]:
    original = child[:]
    child[3] = 1 - child[3]
    f_orig = func(binToDecimal(original))
    f_mut = func(binToDecimal(child))
    if f_mut >= f_orig:
      print(f"{child} accepted")
    else:
      child[3] = 1 - child[3]
      print(f"{child} rejected")
  return offsp
```

```python
def select_survivor(offsp):
  topOffsp = []
  f = getFitness(offsp)
  mInd = f.index(max(f))
  topOffsp.append(offsp[mInd])
  f.pop(mInd)
  mInd = f.index(max(f))
  topOffsp.append(offsp[mInd])
  return topOffsp
```

```python
def best(p):
  f = getFitness(p)
  mInd = f.index(max(f))
  return p[mInd], max(f)
```

```python
population = initPopulation(7,5)
for gen in range(5):
    print(f"Generation {gen+1}")
    print(best(population))
    parents = select_parent(population)
    offsprings = cross(parents)
    offsprings = mutate(offsprings)
    survivors = select_survivor(offsprings)
    population = survivors + population
```
