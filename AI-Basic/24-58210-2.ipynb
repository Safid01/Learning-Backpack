**Before implementation describe the following:**

1. **Agent Description:** The aim of the agent is to act as a smart controller for a WS2812B LED strip. It provides red night-lighting when motion is detected in a dark room, and turns the strip off during the day or when the room is empty to save energy.
2. **List the sensors:** Light Sensor, Motion Sensor
3. **Sensors' values:**  Light -> {'Bright', 'Dark'}
                      Motion -> {'None', 'Detected'}
4. **List the actions:** 'Turn_Red', 'Turn_Off'
5. **Agent function: perception --> action**
   > ['Bright', 'None'] --> 'Turn_Off'
   > ['Bright', 'Detected'] --> 'Turn_Off'
   > ['Dark', 'None'] --> 'Turn_Off'
   > ['Dark', 'Detected'] --> 'Turn_Red'

# 1. Table Driven Agent

```python
table = {
    ('Bright', 'None'): 'Turn_Off',
    ('Bright', 'Detected'): 'Turn_Off',
    ('Dark', 'None'): 'Turn_Off',
    ('Dark', 'Detected'): 'Turn_Red'
}

percepts = []

def table_driven_agent(percept):
    print('Perception Received: ' + str(percept))
    percepts.append(percept)
    action = lookup(percept, table)
    return action

def lookup(p, t):
    return t[p]
```

```python
import time

daily_schedule = [
    ('Bright', 'None'),      # Morning, empty room
    ('Bright', 'Detected'),  # Afternoon, someone walks in
    ('Dark', 'None'),        # Evening, sun sets, room is empty
    ('Dark', 'Detected')     # Night time, someone walks in
]

print("--- Starting Table Driven Agent Simulation ---")
for percept in daily_schedule:
    action = table_driven_agent(percept)
    print('WS2812B Action Performed: ' + action)
    print('-' * 30)
    time.sleep(1)
```

# 2. Simple Reflex Agent

```python
def simple_reflex_agent(percept):
    print('Perception Received: ' + str(percept))
    light_level = percept[0]
    motion_status = percept[1]

    if light_level == 'Dark' and motion_status == 'Detected':
        action = 'Turn_Red'
    else:
        action = 'Turn_Off'

    return action
```

```python
import time

daily_schedule = [
    ('Bright', 'None'),      # Morning, empty room
    ('Bright', 'Detected'),  # Afternoon, someone walks in
    ('Dark', 'None'),        # Evening, sun sets, room is empty
    ('Dark', 'Detected')     # Night time, someone walks in
]

print("--- Starting Simple Reflex Agent Simulation ---")
for percept in daily_schedule:
    action = simple_reflex_agent(percept)
    print('WS2812B Action Performed: ' + action)
    print('-' * 30)
    time.sleep(1)
```
