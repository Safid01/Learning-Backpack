```python
mylist = ['one', 2, 'three', 4, 'five']
```

```python
len(mylist)  # Returns the number of items in the list
```

```python
mylist[1]  # Returns the first item in the list
```

```python
mylist[1:4] # Returns a slice of the list from index 1 to index 3 (not including index 4)
```

```python
my_list = [6, 'seven', 8.1, 'nine', 10]
```

```python
new_list = mylist + my_list  # Concatenates two lists
print(new_list)
```

```python
new_list[7] = 8 # Changes the value at index 7 of new_list to 8
print(new_list)
```

```python
new_list.append('eleven')  # Appends 'eleven' to the end of new_list
print(new_list)
```

```python
new_list.pop()  # Removes the last item from new_list
print(new_list)
```

```python
popped_item = new_list.pop()
print(popped_item)  # Prints the item that was removed
print(new_list)  # Prints the list after the item was removed
```

```python
popped_item = new_list.pop(1)  # Removes the item at index 1 from new_list
print(popped_item)  # Prints the item that was removed
print(new_list)  # Prints the list after the item was removed
```

```python
abcd_list = ['b', 'c', 'a', 'd']
num_list = [4, 3, 1, 2]
```

```python
sorted(abcd_list)  # Returns a new sorted list of abcd_list
```

```python
num_list.sort()  # Sorts num_list in ascending order
print(num_list)
```

```python
num_list.reverse()  # Reverses the order of num_list
print(num_list)
```
